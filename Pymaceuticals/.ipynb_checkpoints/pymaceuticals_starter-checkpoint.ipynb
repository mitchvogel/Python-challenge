## Observations and Insights 

```python
# Analysis observations

# In the Line and Scatter Plots section it is clear that the mouse that I picked had a tumor volume that went down
# as it was treated with Capomulin. This would be expected as it is a successful treatment for tumors. 

# I found the scatter plot of Weight vs Tumor volume to be interesting. It is fair to takeaway from the plot that the
# bigger the mouse is, the bigger the tumor typically is. Further down I calculated the correlation coefficient of 
# thus relationship and returned a value of ~ .84. This is a strong correlation for a data set like this.

# I found the summary statistics table to be interesting. The Capomulin and Ramicane treatments had much lower
# mean and median tumor volumes. One could infer this is due in part to the success of these drug regimens. These two 
# regimens also had much lower variance/st dev. This would mean that the data doesnt fluctuate as much for those two
# drug treatments.

# In a similar vein to the above observation, the IQR of the Capomulin and Ramicane treatments had a significantly
# lower value. This was further evidenced by the box plots which had a smaller "box" or range between bounds. This 
# makes sense given the other data we mined.
```

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
from scipy.stats import linregress

# grab data
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data
study_data = pd.merge(mouse_metadata,study_results,how='outer', on='Mouse ID')
# Display the data
study_data
```

```python
# Checking the number of mice.
mice_count = study_data['Mouse ID'].unique()
og_count = len(mice_count)
og_count
```

```python
mice_count = study_data['Mouse ID'].value_counts()
mice_count
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicates = study_data.set_index('Mouse ID')
g989 = duplicates.loc['g989', :]
# Optional: Get all the data for the duplicate mouse ID. 
g989
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
study_drop = duplicates.drop(index = 'g989')
clean_study_data = study_drop.reset_index()
clean_study_data
```

```python
# Checking the number of mice in the clean DataFrame.
mouse_count_drop = clean_study_data['Mouse ID'].unique()
new_count = len(mouse_count_drop)
new_count
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume.
mean = clean_study_data.groupby('Drug Regimen')['Tumor Volume (mm3)'].mean()
median = clean_study_data.groupby('Drug Regimen')['Tumor Volume (mm3)'].median()
variance = clean_study_data.groupby('Drug Regimen')['Tumor Volume (mm3)'].var()
std_dev = clean_study_data.groupby('Drug Regimen')['Tumor Volume (mm3)'].std()
sem = clean_study_data.groupby('Drug Regimen')['Tumor Volume (mm3)'].sem()
 
# Assemble the resulting series into a single summary dataframe.
data_summary = pd.DataFrame({'Mean':mean,'Median':median,'Variance':variance,'Std Dev':std_dev,'SEM':sem})
data_summary
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Using the aggregation method, produce the same summary statistics in a single line
data_summary_2 = clean_study_data.groupby('Drug Regimen').agg(Mean_Vol=('Tumor Volume (mm3)','mean'),
                                                              Median_Vol=('Tumor Volume (mm3)','median'),
                                                              Var_Vol=('Tumor Volume (mm3)','var'),
                                                              Std_Vol=('Tumor Volume (mm3)','std'),
                                                              SEM_Vol=('Tumor Volume (mm3)','sem')).round(3)
data_summary_2
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
max_timepoint = pd.DataFrame(study_data.groupby('Mouse ID')['Timepoint'].max().sort_values()).reset_index().rename(
                            columns={'Timepoint':'max_timepoint'})
```

```python
# merge max timepoint onto study_data
merged_df = pd.merge(study_data, max_timepoint, on='Mouse ID')
merged_df.head()
```

```python
# Put treatments into a list for for loop (and later for plot labels)
drugs = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']

# Create empty list to fill with tumor vol data (for plotting)
tumor_volume = []

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
for drug in drugs:
    
    # determining Drug Regimen to match drug in for loop
    time_df = merged_df.loc[merged_df['Drug Regimen'] == drug]
    
    # determining point of data at max timepoint
    final_tp_vol = time_df.loc[time_df['Timepoint'] == time_df['max_timepoint']]
    
    # max timepoint volume for drug in loop
    volume = final_tp_vol['Tumor Volume (mm3)']
    
    # add to list
    tumor_volume.append(volume)
    
    # calculate quartiles and iqr
    quartiles = volume.quantile([.25,.5,.75])
    lower_quartile = quartiles[.25]
    upper_quartile = quartiles[.75]
    iqr = upper_quartile - lower_quartile
    
    print(f"IQR for {drug} is {iqr}")
    
    # calculate bounds from quartiles and iqr
    lower_bound = lower_quartile - (1.5*iqr)
    upper_bound = upper_quartile + (1.5*iqr)

                          
    print(f"{drug} potential outliers could be values below {lower_bound} and above {upper_bound}")
    
     # Determine outliers using upper and lower bounds
    outliers = (volume.loc[(final_tp_vol['Tumor Volume (mm3)'] >= upper_bound) | 
                         (final_tp_vol['Tumor Volume (mm3)'] <= lower_bound)]).count()
    
    print(f"Number of {drug} potential outliers is {outliers}")
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest

outlierproperties = dict(marker = '+', markerfacecolor = 'b', markersize = 5, markeredgecolor = 'red')

plt.boxplot(tumor_volume, flierprops = outlierproperties)

plt.title('Final Tumor Volume based on Drug Regimen')
plt.ylabel('Final Volume of Tumor (mm3)')
plt.xticks([1,2,3,4],['Capomulin','Ramicane','Infubinol','Ceftamin'])

plt.show()
```

```python
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen

# grab Capomulin data
capo_treatment = study_data.loc[study_data['Drug Regimen'] == 'Capomulin']

# determine average tumor volume of this subset of data
average_volume = pd.DataFrame(capo_treatment.groupby('Mouse ID')['Tumor Volume (mm3)'].mean().sort_values()).reset_index().rename(columns={'Tumor Volume (mm3)':'average_volume'})

# merge data frames to have average on the Capomulin specific data
average_volume = pd.merge(capo_treatment, average_volume, on='Mouse ID')
cleaned_avg_vol = average_volume[['Weight (g)', 'average_volume']].drop_duplicates()

x_axis = cleaned_avg_vol['Weight (g)']
y_axis = cleaned_avg_vol['average_volume']

plt.scatter(x_axis,y_axis)

plt.xlabel('Weight (g)')
plt.ylabel('Avg Tumor Vol (mm3)')
plt.title('Capomulin Mouse Tumor Volume by Weight')

plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen

# same handful of steps as previous excercise
# grab Capomulin data
capo_treatment = study_data.loc[study_data['Drug Regimen'] == 'Capomulin']

# determine average tumor volume of this subset of data
average_volume = pd.DataFrame(capo_treatment.groupby('Mouse ID')['Tumor Volume (mm3)'].mean().sort_values()).reset_index().rename(columns={'Tumor Volume (mm3)':'average_volume'})

# merge data frames to have average on the Capomulin specific data
average_volume = pd.merge(capo_treatment, average_volume, on='Mouse ID')
cleaned_avg_vol = average_volume[['Weight (g)', 'average_volume']].drop_duplicates()

x_axis = cleaned_avg_vol['Weight (g)']
y_axis = cleaned_avg_vol['average_volume']

# calculate correlation using package
corr = st.pearsonr(x_axis,y_axis)

print(f"The Correlation Coefficient between the weight and average tumor volume for the mouses treated with the Capomulin regimen is {round(corr[0],2)}")

# calculate regression data using package
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_axis,y_axis)
regress = x_axis * slope + intercept
equation = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

plt.scatter(x_axis,y_axis)
plt.plot(x_axis,regress,"r-")

plt.annotate(equation,(22,36),fontsize=10,color='blue')

plt.xlabel('Weight (g)')
plt.ylabel('Avg Tumor Vol (mm3)')
plt.title('Capomulin Mouse Tumor Volume by Weight')
plt.show()
```
